```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Use the existing DataFrame 'df'

# Ensure datetime column is parsed correctly (already done in previous cells, but keeping for completeness)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour

### 1. Heatmap: Traffic volume by hour and location
heatmap_data = df.groupby(['location_id', 'hour'])['traffic_volume'].mean().unstack()
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='YlOrRd', annot=True, fmt=".0f")
plt.title('Traffic Volume Heatmap by Hour and Location ID')
plt.xlabel('Hour of Day')
plt.ylabel('Location ID')
plt.tight_layout()
plt.show()

### 2. Line Plot: Speed trends over time
speed_trend = df.groupby(df['timestamp'].dt.date)['avg_vehicle_speed'].mean()
plt.figure(figsize=(10, 5))
speed_trend.plot(marker='o')
plt.title('Average Vehicle Speed Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Average Vehicle Speed')
plt.grid(True)
plt.tight_layout()
plt.show()

### 3. Scatter Plot: Speed vs Volume
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='traffic_volume', y='avg_vehicle_speed', hue='location_id', alpha=0.6)
plt.title('Average Vehicle Speed vs Traffic Volume by Location ID')
plt.xlabel('Traffic Volume')
plt.ylabel('Average Vehicle Speed')
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
X = df[['avg_vehicle_speed', 'traffic_volume']]
kmeans = KMeans(n_clusters=3, random_state=42).fit(X) # Added random_state for reproducibility
df['cluster'] = kmeans.labels_
```

```python
features = df[['traffic_volume', 'avg_vehicle_speed', 'vehicle_count_cars',
               'vehicle_count_trucks', 'vehicle_count_bikes',
               'temperature', 'humidity']]
```

```python
df_encoded = pd.get_dummies(df[['weather_condition', 'accident_reported', 'signal_status']], drop_first=True)
features = pd.concat([features, df_encoded], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_features)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='traffic_volume', y='avg_vehicle_speed', hue='cluster', palette='Set2')
plt.title('Traffic Clusters: Volume vs Speed')
plt.xlabel('Traffic Volume')
plt.ylabel('Average Speed')
plt.tight_layout()
plt.show()
```

```python
sns.countplot(data=df, x='cluster', hue='location_id')
plt.title('Cluster Distribution by Location')
plt.tight_layout()
plt.show()
```

```python
inertia = []
for k in range(1, 10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(scaled_features)
    inertia.append(km.inertia_)

plt.plot(range(1, 10), inertia, marker='o')
plt.title('Elbow Method for Optimal Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```
