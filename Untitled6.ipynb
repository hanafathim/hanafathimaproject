```python
import pandas as pd
import numpy as np
import logging

# Configure logging
logging.basicConfig(
    filename='data_quality.log',
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s'
)

# Load dataset
try:
    df = pd.read_csv("Final_data.csv")
    logging.info("Dataset loaded successfully.")
except Exception as e:
    logging.error(f"Failed to load dataset: {e}")

# Detect and fill missing values
missing = df.isnull().sum()
logging.info(f"Missing values detected:\n{missing.to_dict()}")

df['Height (m)'].fillna(df['Height (m)'].median(), inplace=True)
df['Weight (kg)'].fillna(df['Weight (kg)'].median(), inplace=True)
logging.info("Missing values filled with median.")

# Remove duplicates
dup_count = df.duplicated().sum()
df.drop_duplicates(inplace=True)
logging.info(f"Removed {dup_count} duplicate rows.")

# Detect and remove outliers
def detect_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    logging.info(f"{len(outliers)} outliers detected in {col}.")
    return outliers

outliers_height = detect_outliers('Height (m)')
outliers_weight = detect_outliers('Weight (kg)')

df = df[~df.index.isin(outliers_height.index)]
df = df[~df.index.isin(outliers_weight.index)]
logging.info("Outliers removed.")

# Validate data types
df['Height (m)'] = pd.to_numeric(df['Height (m)'], errors='coerce')
df['Weight (kg)'] = pd.to_numeric(df['Weight (kg)'], errors='coerce')
logging.info("Data types validated and corrected.")

# Final status
logging.info(f"Final dataset shape: {df.shape}")
```
